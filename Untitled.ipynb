```python

from dotenv import load_dotenv
print(load_dotenv("genote_llm/.env", override=True))

import os
from llama_index.vector_stores import AstraDBVectorStore
from llama_index import (
    VectorStoreIndex,
    SimpleDirectoryReader,
    StorageContext,
)

from llama_index.schema import TextNode, NodeRelationship, RelatedNodeInfo
import openai 




ASTRA_DB_APPLICATION_TOKEN = os.environ.get("ASTRA_DB_APPLICATION_TOKEN")
ASTRA_DB_API_ENDPOINT = os.environ.get("ASTRA_DB_API_ENDPOINT")
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY")

print(OPENAI_API_KEY)

astra_db_store = AstraDBVectorStore(
    token=ASTRA_DB_APPLICATION_TOKEN,
    api_endpoint=ASTRA_DB_API_ENDPOINT,
    collection_name="demo",
    embedding_dimension=1536,
)
storage_context = StorageContext.from_defaults(vector_store=astra_db_store)

text = "The quick brown fox jumps over the lazy dog"



node1 = TextNode(text="The quick brown fox jumps over the lazy dog", id_="<node_id>1")
node2 = TextNode(text="AI", id_="<node_id>2")

nodes = [node1, node2]

index = VectorStoreIndex(
    nodes, storage_context=storage_context
)
```

```python
index = VectorStoreIndex.from_vector_store(vector_store=astra_db_store)


retriever = index.as_retriever(
    vector_store_query_mode="mmr",
    similarity_top_k=1,
)

query_string_2 = "Why did the author choose to work on AI?"
nodes_with_scores = retriever.retrieve(query_string_2)
```
